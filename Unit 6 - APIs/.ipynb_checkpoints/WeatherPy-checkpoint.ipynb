### Analysis
* As expected, the weather becomes significantly warmer as one approaches the equator (0 Deg. Latitude). More interestingly, however, is the fact that the southern hemisphere tends to be warmer this time of year than the northern hemisphere. This may be due to the tilt of the earth.
* There is no strong relationship between latitude and cloudiness. However, it is interesting to see that a strong band of cities sits at 0, 80, and 100% cloudiness.
* There is no strong relationship between latitude and wind speed. However, in northern hemispheres there is a flurry of cities with over 20 mph of wind.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
#Create empty lists
city = []
cloudiness = []
country = []
date = []
humidity = []
latitude = []
longitude = []
temperature = []
windSpeed = []

#Indicate start of data retrieval
print("Beginning Data Retrieval")
print("-----------------------------------")

for i in range(len(cities)):
    
    cityName = cities[i]
    cityURL = cityName.replace(" ", "+")
    
    url = f"http://api.openweathermap.org/data/2.5/weather?units=Imperial&q={cityURL}&appid={api_key}"
    response = requests.get(url).json()
    
    print(f"Processing Record {(i+1)%50} of Set {(i+51)//50} | {cityName}")
    print(url)
    
    try:
        cloudiness.append(response['clouds']['all'])
        country.append(response['sys']['country'])
        date.append(response['dt'])
        humidity.append(response['main']['humidity'])
        latitude.append(response['coord']['lat'])
        longitude.append(response['coord']['lon'])
        temperature.append(response['main']['temp_max'])
        windSpeed.append(response['wind']['speed'])
        city.append(cityName)
        
    except:
        print("City not found. Skipping...")
        pass

    time.sleep(.50)
        
#Indicate end of data retrieval  
print("-----------------------------------")
print("Data Retrieval Complete")
print("-----------------------------------")
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
weather_df = pd.DataFrame({"City": city,
                           "Country": country,
                           "Date": date,
                           "Cloudiness": cloudiness,
                           "Latitude": latitude,
                           "Longitude": longitude,
                           "Humidity": humidity,
                           "Max Temperature": temperature,
                           "Wind Speed": windSpeed})

weather_df.head()
```

```python
weather_df.count()
```

```python
weather_df.to_csv(output_data_file, index = False, header = True)
```

# Linear Regression

```python
#dataframes
northern_df = weather_df.loc[pd.to_numeric(weather_df["Latitude"]).astype(float) > 0, :]
southern_df = weather_df.loc[pd.to_numeric(weather_df["Latitude"]).astype(float) < 0, :]
```

Northern Hemisphere Max Temp vs Latitude



```python
plt.figure(figsize = (15,10))

x_values = pd.to_numeric(northern_df['Latitude']).astype(float)
y_values = pd.to_numeric(northern_df['Max Temperature']).astype(float)
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept

line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

print(f"Regression Equation: {line_eq}")

plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.xlabel('Latitude')
plt.ylabel('Max Temperature (F)')
plt.title('Northern Hemisphere - Max Temp vs. Latitude')

plt.xlim(min(northern_df['Latitude'])-20, max(northern_df['Latitude'])+20)
plt.ylim(min(northern_df['Max Temperature'])-20, max(northern_df['Max Temperature'])+20)

plt.grid(True, color='w', linestyle='-', linewidth=1)
plt.gca().patch.set_facecolor('0.9')

print(f"The r-squared is: {rvalue}")
print("The regression displays a negative correlation.")
print("As you move away from the equator, the temperature decreases.")

plt.savefig("Figures/Figure5.png")

plt.show()
```

# Observations

1. In general, cities that are closer to the equator have higher temperature readings, which hints that latitude and temperatures are correlated. However, it is notable that peak temperatures can be observed at 20 degrees latitude. This can be attributed to the Earth's tilted axis. 
2. Based on the analysis above, there is no discernible relationship between latitude and humidity. Humidity refers to the amount of water vapor in the air. Evaporation is a process that is temperature-dependent, which tells us then that humidity will vary based on temperature. As such, for a more fruitful analysis, it makes more sense to directly investigate the relationship between humidity and temperature, instead of humidity and latitude.
3. Based on the analysis above, there is no discrenible relationship latitude and cloudiness. Cloud formation is dependent on water vapor presence and temperature. As such, similar to humidity, it would make more sense to investigate the relationship between cloudiness and temperature, as opposed to cloudiness and latitude. 
4. There is no discernible pattern that can describe the relationship between wind speed and latitude. Wind speed is affected by temperature, air pressure, earth's rotation, and centripetal acceleration. Since it is not directly affected by latitude, it would make sense that there is no pattern observable when plotting wind speed vs latitude.
5. Linear Regression analysis proves that there is a correlation between latitude and temperature, specifically, in the northern hemisphere: there is a fairly strong negative relationship (-0.88), while in the southern hemisphere: there is a moderate positive relationship (0.67).
6. Linear Regression analysis has also shown that there is little to no correlation between latitude and the three other attributes measured (wind speed, cloudiness, and humidity).
