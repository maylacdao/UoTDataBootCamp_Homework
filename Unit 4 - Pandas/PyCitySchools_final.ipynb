```python
# Dependencies and Setup
import pandas as pd

# Load files
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
data = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Calculate the total number of schools
schools = len(data.school_name.unique())

# Calculate the total number of students
students = data.student_name.count()

# Calculate the total budget
budget = sum(data.budget.unique())

# Calculate the average math score
math_avg = data.math_score.mean()

# Calculate the average reading score
reading_avg = data.reading_score.mean()

# Calculate the percentage of students with a passing math score (70 or greater)
math_passing = ((data[data['math_score'] >= 70].math_score.count()) / students)*100

# Calculate the percentage of students with a passing reading score (70 or greater)
reading_passing = ((data[data['reading_score'] >= 70].reading_score.count()) / students)*100

# Calculate the percentage of students who passed math and reading (% Overall Passing)
overall_passing = (data[(data['math_score'] >= 70) & (data['reading_score'] >= 70)]['student_name'].count()/students)*100

# Create a dataframe to hold the above results
summary_df = pd.DataFrame({"Total Schools":[schools],
                            "Total Students":[students],
                            "Total Budget":[budget],
                            "Average Math Score":[math_avg],
                            "Average Reading Score":[reading_avg],
                            "% Passing Math":[math_passing],
                            "% Passing Reading":[reading_passing],
                            "% Overall Passing Rate":[overall_passing]})

# Optional: give the displayed data cleaner formatting
summary_df['Total Students'] = summary_df.apply(lambda x: "{:,.0f}".format(x['Total Students']), axis=1)
summary_df['Total Budget'] = summary_df.apply(lambda x: "${:,.2f}".format(x['Total Budget']), axis=1)

# Print Summary Dataframe
summary_df
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#function
def avg_math(grade):
    school_summary = data.loc[data.grade == grade].groupby(['school_name'])
    school_names = data.school_name.sort_values().unique()
    school_avg_math = list(school_summary.math_score.mean())
    
    avg_math_df = pd.DataFrame({"School Names": school_names,
                                f"{grade}": school_avg_math})
    
    avg_math_df = avg_math_df.reset_index(drop=True)
    return avg_math_df

#series
grade_9 = avg_math('9th')
grade_10 = avg_math('10th')
grade_11 = avg_math('11th')
grade_12 = avg_math('12th')

#group
avg_math = pd.merge(grade_9, grade_10, how='inner', suffixes=('',''))
avg_math = pd.merge(avg_math, grade_11, how='inner', suffixes=('',''))
avg_math = pd.merge(avg_math, grade_12, how='inner', suffixes=('',''))
avg_math
```

```python
#BINNING FUNCTIONS

def school_averages(df, col):
    groups = df.groupby(col)

    math_avg_df = pd.DataFrame(groups["Average Math Score"].mean()).reset_index()
    reading_avg_df = pd.DataFrame(groups["Average Reading Score"].mean()).reset_index()

    math_passing_df = pd.DataFrame(groups["% Passing Math"].mean()).reset_index()
    reading_passing_df = pd.DataFrame(groups["% Passing Reading"].mean()).reset_index()
    overall_passing_df = pd.DataFrame(groups["% Overall Passing"].mean()).reset_index()

    summary_df = pd.merge(math_avg_df, reading_avg_df, on=col)
    summary_df = pd.merge(summary_df, math_passing_df, on=col)
    summary_df = pd.merge(summary_df, reading_passing_df, on=col)
    summary_df = pd.merge(summary_df, overall_passing_df, on=col)
 
    summary_df = summary_df.set_index(col)
    
    return summary_df

def school_bin(df, col_to_bin, col_to_group, bins, groups):
    df[col_to_group] = df[col_to_group].replace('[\$,)]','', regex=True).astype(float)
    df[col_to_bin] = pd.cut(df[col_to_group], bins, labels=groups)
    summary_df = school_averages(df, col_to_bin)
    summary_df = summary_df.loc[groups,]

    return summary_df
```
